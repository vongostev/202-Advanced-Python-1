```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def data_read(filename):
    file = open(filename, 'r')
    result = [np.empty(0), np.empty(0), np.empty(0), np.empty(0)]
    for line in file:
        line = line.split()
        #print(line[3])
        result[0] = np.append(result[0], float(line[1]))
        result[1] = np.append(result[1], float(line[3])) #300 Hz
        result[2] = np.append(result[2], float(line[5])) #500 Hz
        result[3] = np.append(result[3], float(line[7])) #800 Hz
    return result
```

## Изучение зависимости ЭДС индукции от площади витков катушек

```python
data1 = data_read('data1.txt')
print(data1)
```

```python
#r_1 = find_coeffs(data1_positive, data1_positive[0].size-6)
#err_1 = abs(find_coeffs([data1_positive[0], data1_positive[2]], data1_positive[0].size-6))
x = data1[0]
plt.scatter(x, data1[1], label = '300 Hz')
rms = np.polyfit(data1[0], data1[1], 1)
plt.plot(x, rms[0]*x + rms[1])
print(rms)

plt.scatter(x, data1[2], label = '500 Hz')
rms = np.polyfit(data1[0], data1[2], 1)
plt.plot(x, rms[0]*x + rms[1])
print(rms)

plt.scatter(x, data1[3], label = '800 Hz')
rms = np.polyfit(data1[0], data1[3], 1)
plt.plot(x, rms[0]*x + rms[1])
print(rms)

plt.legend()
plt.xlabel('S, см^2')
plt.ylabel('E, В')
plt.axis([0, 80, 0, 1.5])
plt.grid(True)
plt.title('Зависимость ЭДС индукции в катушке от её площади в зависимости от числа витков')
plt.show()
mnk1 = np.polyfit(data1[0], data1[1], 1)
print(mnk1)
```
