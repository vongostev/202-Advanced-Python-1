```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def func(x, a, b):
    return a*x+b
```

# Бесконтактное Измерение Электропроводности

В данной работе применяется метод, основанный на использовании
комплексной магнитной восприимчивости цилиндрических образцов в
переменном магнитном поле (метод дифференциального трансформатора). Образец 1 помещают внутрь
одной из катушек
трансформатора.
Первичные обмотки
катушек включены
последовательно, и по ним пропускается ток от генератора низкой частоты
(100 – 10000 Гц). Вторичные обмотки включены навстречу друг другу,
поэтому без образца напряжение на выходе дифференциального
трансформатора должно быть равно нулю.

При помещении образца 1 внутрь
рабочей катушки в нем возникают вихревые токи, которые изменяют
магнитное поле, и во вторичной обмотке появляется ЭДС. Так как начальная
ЭДС (без образца) была скомпенсирована второй катушкой, то возникающий
теперь выходной сигнал пропорционален частоте, амплитуде магнитного
поля и эффективной магнитной восприимчивости образца:

$$ U_{out}\infty\frac{\partial{M}}{\partial{t}} = \frac{\partial}{\partial{t}}({\chi H_0} e^{-i\omega t}) = -i \omega \chi H_0 e^{-\omega t} = \omega \chi H_0 e^{-i(\omega t+\frac{\pi}{2}-\beta)} $$

Здесь $\chi=\chi_0 e^{i\beta} $, где $tg\beta=\frac{\alpha"}{\alpha'} $.Иначе говоря,
выходной сигнал оказывается сдвинутым по фазе на величину $\varphi=(\frac{\pi}{2}-\beta)$
относительно магнитного поля (тока в первичной катушке).
Воспользовавшись тем, что $tg(\frac{\pi}{2}-\beta) = ctg\beta $ , получим:


$$tg\varphi = ctg\beta = \frac{\alpha '}{\alpha"}= -\frac{\pi \mu_0 r^2 \sigma}{3}f.$$

Таким образом, построив график зависимости $tg\varphi $ от частоты $f$, по коэффициенту наклона линейного участка кривой можно рассчитать
проводимость $\sigma $.

```python
a=np.loadtxt(open('Cu.txt', 'r'))
x=a[20:,0]
y1=np.tan(a[20:,3]/360*2*3.1415926535)

plt.plot(x,y1, 'bo')
popt, pcov = curve_fit(func, x, y1)

plt.plot(x, func(x, *popt),'r-')

plt.xlabel(r'$f$')

plt.ylabel(r'$tg(\phi)$')

#plt.xlim(100,1100)
#plt.ylim(-4,0)
plt.title(r'$График\;зависимости\;tg(\phi)\;от\;f\;для\;Cu$')
plt.grid()
print("Значения а,b с погрешностями для Сu:")
print("a,b =",popt)
print("Sa,Sb =",np.sqrt(np.diag(pcov)))
plt.savefig('Cu.pdf')

plt.figure()
```

Расчет удельных электропроводностей производится сравнительным
методом. В качестве эталонного образца выбирается медь (σCu = 60 МСм/м).
Тогда в соответствии с соотношением $tg\varphi= -\frac{\pi \mu_0 r^2 \sigma}{3}f$, можно записать для материала
образца i (i = Al, латунь, Ti, Pb и Fe):

$$ \sigma_i=\sigma_{Cu}* \frac{a_i}{a_{Cu}} $$

где $ a_i$ - тангенс угла наклона для материала образца (i), $a_{Cu}$ - тангенс угла
наклона для Cu.

```python
import numpy as np
import matplotlib.pyplot as plt
aCu=-0.00259422
SaCu=3.33153607e-05
aFe=-1.66111972e-04
SaFe=1.96781396e-06
aAl=-0.00129644
SaAl=1.11337864e-05
aL=-0.00130411
SaL=1.89818400e-05
aPb=-0.00029857
SaPb=4.36990021e-06
aTi=-1.38704054e-04
SaTi=2.47334000e-06
sigmaCu=60
SsigmaCu=0

sigmaFe=sigmaCu*aFe/aCu
sigmaAl=sigmaCu*aAl/aCu
sigmaL=sigmaCu*aL/aCu
sigmaPb=sigmaCu*aPb/aCu
sigmaTi=sigmaCu*aTi/aCu
 
SsigmaFe=(sigmaCu/abs(aCu))*((SaFe**2+(aFe*SaCu/aCu)**2)**0.5)
SsigmaAl=(sigmaCu/abs(aCu))*((SaAl**2+(aAl*SaCu/aCu)**2)**0.5)
SsigmaL=(sigmaCu/abs(aCu))*((SaL**2+(aL*SaCu/aCu)**2)**0.5)
SsigmaPb=(sigmaCu/abs(aCu))*((SaPb**2+(aPb*SaCu/aCu)**2)**0.5)
SsigmaTi=(sigmaCu/abs(aCu))*((SaTi**2+(aTi*SaCu/aCu)**2)**0.5)


print("sigmaFe=",sigmaFe, "+-", SsigmaFe)
print("sigmaAl=",sigmaAl, "+-", SsigmaAl)
print("sigmaL=",sigmaL, "+-", SsigmaL)
print("sigmaPb=",sigmaPb, "+-", SsigmaPb)
print("sigmaTi=",sigmaTi, "+-", SsigmaTi)
```
